```python
import pandas as pd
```

```python
file = 'Wording_Table2017ALL_V1.xlsx'
```

```python
xl = pd.ExcelFile(file)
```

```python
print(xl.sheet_names)
```

```python
df1 = xl.parse('TEST')
```

## 填補空值

```python
df2 = df1.fillna(0)
```

```python
from sklearn.ensemble import (RandomTreesEmbedding, RandomForestClassifier,
                              GradientBoostingClassifier)

from sklearn.model_selection import train_test_split
```

## 切割資料集

```python
y = df2.iloc[:, 225]
X = df2.iloc[:, 2:225]
#將資料集切出特徵值X  以及目標值y
```

```python
X
```

```python
X_train, X_test, y_train, y_test  = train_test_split(X, y, test_size=0.2)
```

```python
X_test
```

## 以GDBT演算法預測漲跌

```python
params = {'n_estimators': 30, 'max_depth':3, 'learning_rate' :0.01}

clf = GradientBoostingClassifier(**params)
clf.fit(X_train, y_train)
clf.predict(X_test)
```

```python
clf.score(X_test, y_test)
#測試GDBT精確度
```

```python
elf = RandomForestClassifier(n_estimators=40, max_depth=5, random_state=10)
elf = elf.fit(X,y)
elf.predict(X_test)
```

```python
elf.score(X_test, y_test)
```

```python
from sklearn.svm import SVC
dlf = SVC(kernel='rbf', probability=True)
dlf = dlf.fit(X,y)
dlf.predict(X_test)
```

```python
dlf.score(X_test, y_test)
```

```python
from sklearn.ensemble import VotingClassifier
```

```python
flf = VotingClassifier(estimators=[('GradientBoostingClassifier', clf), ('RandomForestClassifier', elf), ('svc', dlf)], voting='soft', weights=[2,1,2])
```

```python
flf = flf.fit(X,y)
```

```python
flf.predict(X_test)
```

```python
flf.score(X_test, y_test)
```

```python
from numpy import loadtxt
from xgboost import XGBClassifier
from xgboost import plot_tree
import matplotlib.pyplot as plt
# load data

# fit model no training data
llf = XGBClassifier()
llf.fit(X, y)
llf.score(X_test, y_test)
```

```python
llf.predict(X_test)
```

```python
plot_tree(llf, num_trees=10, rankdir='LR')
plt.show()
```

```python
from sklearn.model_selection import cross_val_score
for lf, label in zip([clf, elf, dlf, flf, llf], ['GradientBoostingClassifier', 'RandomForestClassifier', 'svc', 'VotingClassifier', 'XGBoost']):
    scores = cross_val_score(lf, X, y, cv=20, scoring='accuracy')
    print("Accuracy: %0.4f (+/- %0.2f) [%s]" % (scores.mean(), scores.std(), label))
```

## 柱狀圖(預測及實際漲跌狀況)

```python
t = list(y_test)
r = clf.predict(X_test) #GBDT
f = dlf.predict(X_test) #RandomForest
g = elf.predict(X_test) #SVM
m = flf.predict(X_test) #Voting
o = llf.predict(X_test)   #XGBOOST
```

## 1. GBDT

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import pyplot as plt
plt.figure(figsize=(35,9))
n = 35
X = np.arange(n)+1
#X是1-35 testing data柱狀圖的個數

objects = range(1,36)

#設定x軸1-35柱狀圖的標籤

y_pos = np.arange(len(objects))

#柱狀圖x軸每隔1個給1個標籤

plt.bar(X,t,width = 0.3,facecolor = 'lightskyblue',edgecolor = 'white')

#width:柱的寬度, t為測試資料的目標值

plt.bar(X+0.3,r,width = 0.3,facecolor = 'yellowgreen',edgecolor = 'white')

#水平柱狀圖plt.bar，屬性中寬度width變成了高度height
#打兩組數據時用+
#facecolor柱狀圖裡填充的顏色
#edgecolor是邊框的顏色
#想把一組數據打到下邊，在數據前使用負號
#plt.bar(X, -Y2, width=width, facecolor='#ff9999', edgecolor='white')
#給圖加text

plt.xlabel('Data')
plt.ylabel('Actual(blue)   Prediction(green)')

#給x軸和y軸加上label

plt.xticks(y_pos+1, objects)

#每組柱狀圖給標籤在上面

plt.title('GDBT')

#標題GDBT

for x,y in zip(X,t):
    plt.text(x, y, '%.2f' % y, ha='center', va= 'bottom')

for x,y in zip(X,r):
    plt.text(x, y, '%.2f' % y, ha='center', va= 'bottom')
    
plt.ylim(-1,1)
#y軸範圍

plt.show()
```
